```python
import numpy as np
```

```python
marks = [[4,2,3,1], [5,9,4,0], [1,9,3,5]] #consider this a matrix of marks of each student in different subjects,
```

```python
np.array(marks) # maths, biology, physics, chemistry
```

```python
marks[0] #this gave me the marks of one student in all subjects
```

```python
marks[0][0], marks[1][0], marks[2][0] #this gave me the marks of students in maths only
```

```python
#this is tedious, so we transpose the matrix and now we have subject first matrix, so we can extract the marks of all students in a particular subject
#This is the use of transpose.

```

```python
#traditional method(using loops)

transpose_matrix=[] #make an empty matrix
for i in range(len(marks[0])):
  transpose_matrix.append([]) #adds an empty list here
  for j in range(len(marks)):
    transpose_matrix[i].append(marks[j][i])

transpose_matrix
```

```python
#using numpy
marks_array = np.array(marks)
transposed = marks_array.T
```

```python
# marks_array = subjectwise marks for each student
# transposed = marks in a subject per student

#so if im a math teacher who wants to calculate class avg, i will need the second matrix,
transposed.sum(axis=1)/3
```

```python
#the above code gave avg marks in each subject.
```

```python
marks_array.sum(axis=1)/40*100    #4 subjects of 10 marks each, so this gives percentage of each student
```

```python
#axis=1 performs row wise operation, axis=0 performs column wise operation, so transpose was not really needed here, but it has other use cases which are not reflected here
```
