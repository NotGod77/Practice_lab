```python
import numpy as np
```

```python
#adding different dimensional matrices, is called broadcasting

# [ [ 1  2  3  4 ]       [ 1 ]       [ [ 1  2  3  4 ]       [ 1  1  1  1 ]
#   [ 5  6  7  8 ]   +   [ 2 ]   =     [ 5  6  7  8 ]   +     [ 2  2  2  2 ]
#   [ 9 10 11 12 ]       [ 3 ]         [ 9 10 11 12 ]         [ 3  3  3  3 ] ]
```

```python
#this is called broadcasting of a vector.

#the only condition is that the number of columns in the matrix to be broadcasted has to be 1, or it should just be a vector
#And the number of rows of both should be same.
#for both matrices of orders MxN and PxQ, (M=P and Q=1)
```

```python
marks = np.random.randint(6,size=(3,4))
marks_array=np.array(marks) #convert to a numpy array for operations using numpy.
marks #same, consider the rows to be marks of a student in 4 subjects.
```

```python
grace_marks=np.array([[1],[2],[3]]) #suppose i want to give 1 mark to S1, 2 marks to S2 and 3 marks to S3 as grace
#so this is a 2d array of order 3x1. This will do.
marks_array+grace_marks
```

```python
#we see that 1, 2 and 3 have been successfully broadcasted to each row
#broadcasting just needs you to add the matrices with certain conditions, numpy will do the hardwork for ya
```
