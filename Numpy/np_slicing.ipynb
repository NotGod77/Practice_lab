```python
int_arr = np.random.randint(6,size=(6,8))  # make a random integer array
int_arr
```

```python
row = int_arr[0]   # assign one row of the 2d array to a variable
row
```

```python
boolean_arr = (row == 0) | (row == 1) | (row == 2)  # comparing list directly gives a boolean array.
```

```python
#now our main goal is not to know whether the element is present or not, our goal is to select the true elements and discard false ones.
#we do that by using the boolean array as a mask.
boolean_arr #this is the array
```

```python
row[boolean_arr] #original array[boolean array] -> this gives only the true elements,
```
