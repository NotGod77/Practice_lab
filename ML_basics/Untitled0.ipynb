```python
import pandas as pd
from sklearn import linear_model
df = pd.read_csv('sample_data/california_housing_test.csv')

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['median_house_value'], bins=50, edgecolor='black')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.title('Distribution of Median House Value')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df.median_house_value,df.median_income,alpha=0.5)
plt.xlabel('Median House Value')
plt.ylabel('Median Income')
plt.title('House Value vs Income')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['median_income']], df.median_house_value)
#here, median income is the independent var, and the second one is dependent, so we are training the model to predict the second value using the first value.
#As income changes, how does the house value change?
```

```python
# Predict the median house value for a median income of 5.0
predicted_value = reg.predict([[5.0]])
print(f"Predicted median house value for a median income of 5.0: {predicted_value[0]:.2f}")
```
