```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
data = {
    "Name": [
        "Rob","Michael","Mohan","Ismail","Kory","Gautam","David","Andrea","Brad","Angelina",
        "Donald","Tom","Arnold","Jared","Stark","Ranbir","Dipika","Priyanka","Nick","Alia","Sid","Abdul"
    ],
    "Age": [
        27,29,29,28,42,39,41,38,36,35,
        37,26,27,28,29,32,40,41,43,39,41,39
    ],
    "Income($)": [
        70000,90000,61000,60000,150000,155000,160000,162000,156000,130000,
        137000,45000,48000,51000,49500,53000,65000,63000,64000,80000,82000,58000
    ]
}

df = pd.DataFrame(data)
df
```

```python
plt.scatter(df.Age,df['Income($)'])
```

## Clustering

```python
# we can make three clusters
km= KMeans(n_clusters=3)
cluster_number = km.fit_predict(df[['Age','Income($)']]) #divides each data point into clusters
cluster_number
```

```python
df['Cluster_no'] = cluster_number
df
```

```python
plt.scatter(df.Age,df['Income($)'],c=df['Cluster_no'],cmap = 'viridis')
```

## Scaled Clustering

# New Section

```python
#Here our clusters are somewhat not made correctly, because our data is not scaled,
scaler = MinMaxScaler()
scaler = MinMaxScaler()

df['Income($)'] = scaler.fit_transform(df[['Income($)']])
df['Age'] = scaler.fit_transform(df[['Age']])
df
```

```python
km = KMeans(n_clusters=3)
df['Clus_no']=km.fit_predict(df[['Age','Income($)']])
df
```

```python
km.cluster_centers_
```

```python
plt.scatter(df.Age,df[['Income($)']],c=df.Clus_no)#Now we can see the clusters are well made
# let's even plot the centroids of the clusters
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color= 'purple',marker = 'x',label = 'centroid')
```

## Elbow plot

```python
#Sometimes the number of clusters you have to take is not clear, so we can plot different number of clusters with their SSE, and loook for and elbow

sse=[]
k_rng = range(1,11)
for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(df[['Age','Income($)']])
  sse.append(km.inertia_)
```

```python
plt.plot(k_rng,sse)
```

```python
# we see an elbow at 3, so 3 is the best number of clusters
```
