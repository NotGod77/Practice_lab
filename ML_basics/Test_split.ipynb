#Splitting data into Training and testing sets

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = {
    'Travelled': [69000, 35000, 57000, 22500, 46000, 59000, 52000, 72000, 91000, 67000,
                83000, 79000, 59000, 58780, 82450, 25400, 28000, 69000, 87600, 52000],
    'Age': [6, 3, 5, 2, 4, 5, 5, 6, 8, 6,
                 7, 7, 5, 7, 7, 3, 2, 5, 8, 5],
    'price': [18000, 34000, 26100, 40000, 31500, 26750, 32000, 19300, 12000, 22000,
                      18700, 19500, 26000, 27500, 19400, 35000, 35500, 19700, 12800, 28200]
}

df = pd.DataFrame(data)
```

```python
plt.scatter(df.Age,df.price)
```

```python
plt.scatter(df.Travelled,df.price)
```

```python
X = df[['Travelled','Age']]
y = df[['price']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.1)
```

```python
len(x_train)
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.predict(x_test)
```

```python
y_test
```

```python
reg.score(x_test,y_test)
```
