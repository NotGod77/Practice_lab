```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

#Data preprocessing

```python
df =pd.read_csv('/content/sample_data/mult_var.csv')
df
```

```python
import math
filling_num=math.floor(df.bedrooms.median())
```

```python
df.bedrooms = df.bedrooms.fillna(filling_num)
```

```python
df
```

##Train Linear model

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```

```python
reg.predict([[3000,4,5]])
```
